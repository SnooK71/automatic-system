```
from __future__ import absolute_import, division, print_function, unicode_literals

# Install TensorFlow

import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd

print(tf.__version__)
```

```
url = 'https://raw.githubusercontent.com/SnooK71/automatic-system/master/71_clear.csv'
df = pd.read_csv(url, sep=',', encoding="cp1252")
```

```
df = df.drop(['numero','adresse_numero','adresse_nom_voie','id_parcelle','nombre_lots','code_type_local','code_nature_culture','nom_commune'], axis=1)
```

```
df.head()
```

```
df['Mean'] = 0
```

```
df.head()
```

```
print(df['type_local'].value_counts())
```

```
df.columns
#data_macon = print(df.iloc[(df['code_postal'].values==71000.),:])
#data_macon = print(df.loc[(df['type_local'].values=="Appartement"),:])
data_macon = df.iloc[(df['code_postal'].values==71000.) & (df['type_local'].values=="Appartement"),:]
data_macon
```

```
data_macon = data_macon.drop(['type_local'], axis=1)
data_macon
```

```
dataset = data_macon.values
dataset
```

```
X = dataset[:,0:6]
Y = dataset[:,6]
```

```
from sklearn import preprocessing
```

```
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```
X_scale
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.2)
```

```
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```
model = Sequential([
    Dense(32, activation='relu', input_shape=(6,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```
model.fit(X_train, Y_train, batch_size=32, epochs=100, validation_data=(X_val, Y_val))
```

```
model.evaluate(X_test, Y_test)
```
